```python
import numpy as np
import matplotlib as mpl
#mpl.__version__
import matplotlib.pyplot as plt # Import the main plotting (sub) package with the usual abbreviation plt.
plt.style.use('seaborn')  # Sets the plotting style to seaborn
mpl.rcParams['font.family'] = 'serif' # Sets the font to be serif in all plots
%matplotlib inline
```

```python
strike = np.linspace(50, 150, 24) # The ndarray object with the strike values
ttm = np.linspace(0.5, 2.5, 24) # The ndarray object with the time-to-maturity values
strike, ttm = np.meshgrid(strike, ttm)  # The two 2D ndarray objects (grids) created
strike[:2].round(1)
```

```python
iv = (strike - 100) ** 2 / (100 * strike) / ttm # The dummy implied volatility values
iv[:5, :3]
```

```python
# Imports the relevant 3D plotting features, which is required although Axes3D is not directly used
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(projection = '3d') # Sets up a canvas for 3D plotting
surf = ax.plot_surface(strike, ttm, iv, rstride=2, cstride=2,
                       cmap=plt.cm.jet, linewidth=0.5,
                       antialiased=True)  # Creates the 3D plot
ax.set_xlabel('strike') # Sets the x-axis label
ax.set_ylabel('time0to-maturity') # Sets the y-axis
ax.set_zlabel('implied volatility') # Sets the z-axis
fig.colorbar(surf, shrink=0.5, aspect=5)  # Creates a colour bar
```

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.view_init(30, 60)
ax.scatter(strike,ttm, iv, zdir='z', s=25,
           cmap='b', marker='^')
ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility');
```
